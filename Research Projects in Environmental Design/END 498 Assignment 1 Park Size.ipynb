```python
from sympy import *
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
#plt.style.use('fivethirtyeight')
#plt.style.use('bmh')
#plt.style.use('seaborn-whitegrid')

# Set up data
fig = plt.figure()
x = fig.add_axes([0,0,1,1])
hood = ['Poor', 'Wealthy', 'Black','White']
acres = [6.4, 7.9, 14, 29.8]
width = 0.5
color = ['blue']

# Plot the motherfucker
x.bar(hood, acres, width, color=color)
x.set_title('Average Park Sizes Between Different Neighborhood Classes', fontsize = 16)
x.set_xlabel('Neighborhoods', fontsize = 15)
x.set_ylabel('Acres', fontsize = 15)
x.set_axisbelow(True)

# Adjust tick sizes, for graphical effects
Axes.tick_params(x, axis = 'x', labelsize = 13)
Axes.tick_params(x, axis = 'y', labelsize = 13)

# Display the plot and save it
figsize(8,5)
plt.grid(b = True, which = 'major', axis = 'both')
plt.savefig('Park Size.png', bbox_inches = 'tight')
plt.show()
```
