# Statistical Analysis

It follows that the histogram resemble a normal distribution with $\mu$=66.2 and $\sigma^2$=131.1.
We will now attempt to verify our expectation and variance by computing the maximum likelihood estimator for our Walk Score. That is, we will determine the function that is the most likely to have generated this data and be certain that this is the function. Doing so will allow us to perform a wide range of statistical analysis.

The probability density function of a normal distribution is 
$$ f_{x}(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} $$
Plugging in our numbers yield:
$$ \begin{align}
f_{x}(x) &= \frac{1}{131.0988\sqrt{2\pi}} e^{-\frac{(x-66.2)^2}{2*131.0988^2}} \\
f_{x}(x) &= \frac{1}{328.616} e^{-\frac{(x-66.2)^2}{34373.7907}} 
\end{align}$$

---

$f_x (x)$ is the function that generated our data. We will now attempt to prove that this is the function that most likely to have generated this data. We will do this by computing the maximum likelihood estimator for our parameters $\mu$ and $\sigma^2$. That is, we will compute the MLE as if we do not have $\mu$ and $\sigma^2$. 

Let $X_i$ for $i = 1, 2,...,n$ be n-independent and identically distributed random variables with pmf $f(x;\theta)$. Given a data set $x_1, x_2, ... , x_n$, we define the likelihood function as:
$$L(x_1 , ... , x_n; \mu,\sigma) = \prod_{i=1}^{n}f(x;\theta)$$
Adapting this function to our purposes yield the following liklihood function for our Walk Score:
$$
\begin{align}
L(44, 44, ..., 94; \mu,\sigma) &= \prod_{i=1}^{54}f(x;\mu,\sigma^2) \\
L(44, 44, ..., 94; \mu,\sigma ) &= \prod_{i=1}^{54} \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x_i -\mu)^2}{2\sigma^2}}
\end{align}
$$

To determine the maximum of any function, we would differentiate the function with respect to our parameter once and set it equal to 0. In this case, we would be differentiating a product of a function 54 times. Our method for differentiating product of functions, the product rule, will take eons to complete, and the derivative of a product is not equal to the product of its derivatives. Therefore, we will take the natural log of our likelihood function to reduce it to a summation of functions. Doing so will allow us to easily take the derivative, since the derivative of a sum is equal to the sum of its derivative.

$$
\begin{align}
ln\left(L(x_1 = 44, ..., x_{54} = 94; \mu,\sigma)\right) &= ln\left(\prod_{i=1}^{54} \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x_i -\mu)^2}{2\sigma^2}}\right) \\
&= \sum_{i=1}^{54} ln\left(\frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x_i -\mu)^2}{2\sigma^2}}\right) \\
&= \frac{n}{\sqrt{2\pi}} \sum_{i=1}^{54} ln\left(\frac{1}{\sigma} e^{-\frac{(x_i -\mu)^2}{2\sigma^2}}\right) \\
&= \frac{n}{\sqrt{2\pi}} \sum_{i=1}^{54} ln\left(\frac{1}{\sigma}\right) + \sum_{i=1}^{54} ln\left(e^{-\frac{(x_i -\mu)^2}{2\sigma^2}}\right) \\
&= \frac{n}{\sqrt{2\pi}} \sum_{i=1}^{54} ln\left(\frac{1}{\sigma}\right) + \sum_{i=1}^{54} \left(-\frac{(x_i -\mu)^2}{2\sigma^2}\right) \\
&= \frac{n}{\sqrt{2\pi}} \sum_{i=1}^{54} ln\left(\frac{1}{\sigma}\right) - \sum_{i=1}^{54} \left(\frac{(x_i -\mu)^2}{2\sigma^2}\right) \\
\end{align}
$$

Our new likelihood function has been reduced as much as we could. We will now take the partial derivative with respect to $\mu$ and then $\sigma$.

$$
\begin{align}
\frac{\partial ln(L)}{\partial \mu} &= \frac{\partial}{\partial \mu}\frac{n}{\sqrt{2\pi}} \sum_{i=1}^{54} ln\left(\frac{1}{\sigma}\right) - \frac{\partial}{\partial \mu} \sum_{i=1}^{54} \left(\frac{(x_i -\mu)^2}{2\sigma^2}\right) \\
&= 0 - \frac{\partial}{\partial \mu} \sum_{i=1}^{54} \left(\frac{(x_i -\mu)^2}{2\sigma^2}\right) \\
&= -\frac{1}{2\sigma^2} \sum_{i=1}^{54} \frac{\partial}{\partial u}(u)^2 \frac{\partial u}{\partial \mu} x_i -\mu \\
&= -\frac{1}{2\sigma^2} \sum_{i=1}^{54} 2u * -1 \\
&= -\frac{1}{2\sigma^2} \sum_{i=1}^{54} 2(x_i -\mu) * -1 \\
&= -\frac{1}{2\sigma^2} \sum_{i=1}^{54} -2(x_i -\mu) \\
&= -\frac{1}{2\sigma^2} -2\sum_{i=1}^{54} x_i -\mu \\
&= \frac{1}{\sigma^2} \sum_{i=1}^{54} x_i -\mu \\
\end{align}
$$

We will now set the partial derivative of the natural log of our likelihood function equal to 0.

$$
\begin{align}
\frac{1}{\sigma^2} \sum_{i=1}^{54} x_i - \mu &= 0 \\
\frac{1}{\sigma^2} \sum_{i=1}^{54} x_i - \sum_{i=1}^{54}\mu &= 0 \\
\frac{1}{\sigma^2} \sum_{i=1}^{54} x_i - n\mu &= 0 \\
\sum_{i=1}^{54} x_i &= n\mu \\
\frac{1}{n} \sum_{i=1}^{54} x_i &= \hat{\mu} \\
\bar{x} = \hat{\mu} \\
\end{align}
$$

We have thus proved that the most likely function to estimate our mean is by taking the average. Not only that, but $\hat{\mu}$ is the Uniformly Minimum Variance Unbiased Estimator of $\hat{\theta}$. Meaning that we have the best estimator of our sample mean of 66.2.

___

With walk scores of 94 and 48, we will perform hypothesis testing to determine how rare the walk score average is. Since we are dealing with $\mu$, and the population standard deviation $\sigma$, we will employ the Z-test with $\alpha = 0.05$.

The test statistic for a z-distrbution is the following pivotal quantity:

$$z = \frac{\bar{x} - \mu_0}{\sigma}$$

Plugging in our walk scores yield the following:

$$
\begin{align}
z &= \frac{48 - 66.2}{11.45} \\
z &= \frac{-18.2}{11.45} \\
\therefore z &= -1.5915 \\
\end{align}
$$

According to our test statistic, Grider's walk score is about 1.6 standard deviations to the left of the mean, which explains the negative value. We now determine the Cumulative Distribution Function, or CDF, at 48. That way, we will know how much probability has accumulated up to and including 48. Ths tells us how rare or how often it is for a walk score to be under that number, or $P(x \in [0,100] < \bar{X})$.

We will now set up the critical region for Grider, or the the area that a walk score has to be in orer for it to be considered rare. Since Grider's test statistic is a negative value, or in other words, since Grider is below $\mu$, we will compute the left-sided critical region for $f_{x}(x)$.

For our significance level (or $P\big(\text{Test Statistic} \in \text{Critical Region} | \text{$H_o$ is true}\big)$), it will be set at 0.05, or $\alpha \leq 0.05$. We have the population mean, $\mu$, population standard deviation, $\sigma$, and we are dealing with the population mean, $\mu$, so we will compute $Z_{0.05}$.

In order to compute the value such that the area under the distribution is 0.05, we need to utilize the t-distribution. As $\nu \to \infty$, the t-distribution closely resembles a z-distribution, which will help us compute $Z_{0.05}$. We will compute $\int_{0}^{\infty}f_{t_\nu}(t)dt$, where the t-distribution, $f_{t_\nu}(t) = \frac{\Gamma\big[\frac{\nu + 1 }{2}\big]}{\sqrt{\nu\pi}\Gamma\frac{\nu}{2}} \Big(1 + \frac{t^{2}}{\nu} \Big)^{\frac{-(\nu+1)}{2}}.$ As direct integration of $f_{t_\nu}(t)$ is impossible, numerical integration will be used to approximate the values of $\nu$ by employing the age-old technique of a table of values using Microsoft Excel. Due to the nature of Excel and infinity, we will numerically compute $\int_{0}^{9999}f_{t_\nu}(t)dt$.

According to Excel, $Z_{0.05} \approx 1.645006$. This will be our critical region for Grider and, in fact, all walk scores below $\mu$.

Transforming our test statistic into our data values, we get the following:
$$
\begin{align}
\mu_0 * z &= 66.\bar{2} - (1.645006*11.45) \\
&= 47.3872
\end{align}
$$

What this means is that our critical region will begin at 47.3872. Our walk score for Grider is 48, which means it is not in the critical region, which means a walk score of 48 is not rare for our data. In fact, Grider's walk score is 0.6128 units away from being considered rare, so very close to the critical region.

Let's compute the p-value for Grider. Doing so will tell us the probability of getting a walk score that's smaller than Grider

Let's compute power, or $P(\text{Reject} H_o | H_a \text{is true})$

For Allentown's walk score, it is to the right of $\mu$, so we will compute the right-sided critical region for $f_{x}(x)$.

Test statistic:
$$
\begin{align}
z &= \frac{94 - 66.2}{11.45} \\
z &= \frac{27.\bar{7}}{11.45} \\
\therefore z &= 2.4260 \\
\end{align}
$$


Note that 2.4260 only tells us how many standard deviations away our walk score is from $\mu$. We will use the actual walk score to determine its location relative to the critical region, which is:
$$
\begin{align}
\mu_0 * z &= 66.\bar{2} + (1.645006*11.45) \\
&= 85.0573
\end{align}
$$

Our right-sided critical region will begin at 85.0573. Since Allentown's walk score is 94, the number is well into the critical region, making Allentown's walk score rare.
