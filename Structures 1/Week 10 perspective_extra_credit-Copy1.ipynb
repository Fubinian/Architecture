According to the interactive plot, the corners of the billboard have the following coordinates:

|    <i></i>      |  **x**  |  **y**  |
|-----------------|---------|---------|
| **Top**         | 294.397 | 511.738 |
| **Right**       | 303.990 | 592.109 |
| **Left**        | 221.545 | 563.590 |
| **Bottom**      | 375.287 | 539.479 |

This means that the camera coordinates $\mathcal{C}$ are:

$[w_{1}]_\mathcal{C} = \begin{bmatrix} 1 \\ 294.397 \\ 511.738 \end{bmatrix}$, $\hspace{2mm}$ 
$[w_{2}]_\mathcal{C} = \begin{bmatrix} 1 \\ 303.990 \\ 592.109 \end{bmatrix}$, $\hspace{2mm}$ 
$[w_{3}]_\mathcal{C} = \begin{bmatrix} 1 \\ 221.545 \\ 563.590 \end{bmatrix}$, $\hspace{2mm}$ 
$[w_{4}]_\mathcal{C} = \begin{bmatrix} 1 \\ 375.287 \\ 539.470 \end{bmatrix}$ $\hspace{2mm}$ 

Camera coorinates $\mathcal{C}$ entered into Python code:

```python
A = Matrix([[1],[294.397],[511.738]]) #[w_1]_C
B = Matrix([[1],[303.990],[592.109]])#[w_2]_C
C = Matrix([[1],[221.545],[563.590]])#[w_3]_C
D = Matrix([[1],[375.287],[539.470]]) #[w_4]_C

A,B,C,D
```

If $m_1, m_2, m_3$,and $m_4$ are the lengths of the mural, and $a, b, c$, and $d$ are the mural coordinates, then 

$$m_1 = a*w_1 \tag{1}$$

$$m_2 = b*w_2 - a*w_1 \tag{2}$$

$$m_3 = c*w_3 - a*w_1 \tag{3}$$

So, if $m_1 + m_2 + m_3 = d*w_4$, then

$$b*w_2 + c*w_3 - a*w_1 = d*w_4 \tag{4}$$

$$b*[w_2]_{\mathcal{C}} + c*[w_3]_{\mathcal{C}} - a*[w_1]_{\mathcal{C}} = d*[w_4]_{\mathcal{C}} \tag{5} $$

$$b*\begin{bmatrix} 1 \\ 303.990 \\ 592.109 \end{bmatrix} + c*\begin{bmatrix} 1 \\ 221.545 \\ 563.590 \end{bmatrix} - a*\begin{bmatrix} 1 \\ 294.397 \\ 511.738 \end{bmatrix} = d*\begin{bmatrix} 1 \\ 375.287 \\ 539.470 \end{bmatrix} \tag{6}$$

Setting $d = 1$ leaves us with

$$b*\begin{bmatrix} 1 \\ 303.990 \\ 592.109 \end{bmatrix} + c*\begin{bmatrix} 1 \\ 221.545 \\ 563.590 \end{bmatrix} - a*\begin{bmatrix} 1 \\ 294.397 \\ 511.738 \end{bmatrix} = \begin{bmatrix} 1 \\ 375.287 \\ 539.470 \end{bmatrix}\tag{7}$$

Distributing the minus (-) sign in front of $a$:

$$b*\begin{bmatrix} 1 \\ 303.990 \\ 592.109 \end{bmatrix} + c*\begin{bmatrix} 1 \\ 221.545 \\ 563.590 \end{bmatrix} + a*\begin{bmatrix} -1 \\ -294.397 \\ -511.738 \end{bmatrix} = \begin{bmatrix} 1 \\ 375.287 \\ 539.470 \end{bmatrix} \tag{8}$$

What we have here is a vector equation. Standard procedures for solving a vector equation will be followed.

$$\begin{bmatrix} 1 & 1 & -1 & 1 \\ 303.990 & 221.545 & -294.397 & 375.287\\ 592.109 & 563.590 & -511.738 & 539.470 \end{bmatrix} \tag{9}$$

```python
OBC = Matrix([[1,1,-1,1],[303.990,221.545,-294.397,375.287],[592.109,563.590,-511.738,539.470]])
OBC
```

Simplifying the augmented matrix

```python
Matrix([[1,0,0,0.978],[0,1,0,-0.982],[0,0,1,-1.003]])
```

Verifying the soutions to the equation

```python
0.978282232470878*B + -0.98151510657095*C - -1.00323287410007*A
```

This vector is equal to $[w_4]_{\mathcal{C}}$. Therefore, we obtain the solution to the vector equation in $(8)$

$$\begin{cases}  b = 0.978 \\ c = −0.982 \\ a = −1.003  \end{cases} \Longrightarrow \begin{cases} a = −1.003 \\  b = 0.978 \\ c = −0.982 \end{cases} \tag{10}$$ 

Which means

$$[v]_\mathcal{B} = \begin{bmatrix} −1.003 \\ 0.978 \\ −0.982 \end{bmatrix} \tag{11}$$

For the mural coordinate system $[m]_{\mathcal{C}}$, we have, for $[m_{1}]_{\mathcal{C}}$:

$$m_1 = a*w_1 \longrightarrow [m_1]_\mathcal{C} = a*[w_1]_\mathcal{C} \tag{12}$$

which gives

$$[m_1]_{\mathcal{C}} = −1.003 * \begin{bmatrix} 1 \\ 294.397 \\ 511.738 \end{bmatrix} \tag{13}$$

```python
-1.003*Matrix([[1],[294.397],[511.738]])
```

$$\therefore [m_1]_{\mathcal{C}} = \begin{bmatrix} −1.003 \\ −295.280191 \\ −513.273214 \end{bmatrix} \tag{14}$$

For $[m_{2}]_{\mathcal{C}}$, we have

$$m_{2} = b*w_{2} - a*w_{1} \longrightarrow [m_2]_{\mathcal{C}} = b*[w_2]_{\mathcal{C}} - a*[w_1]_{\mathcal{C}} \tag{15}$$

$$[m_2]_{\mathcal{C}} = 0.978*\begin{bmatrix} 1 \\ 303.990 \\ 592.109 \end{bmatrix} - -1.003* \begin{bmatrix} 1 \\ 294.397 \\ 511.738 \end{bmatrix} \tag{16}$$

```python
0.978*Matrix([[1],[303.990],[592.109]]) - -1.003*Matrix([[1],[294.397],[511.738]])
```

$$\therefore[m_2]_\mathcal{C} = \begin{bmatrix} 1.981 \\ 592.582411 \\ 1092.355816 \end{bmatrix} \tag{17}$$

Finally, for $[m_{3}]_{\mathcal{C}}$, we have

$$m_3 = c*w_3 - a*w_1 \longrightarrow [m_3]_{\mathcal{C}} = c*[w_3]_{\mathcal{C}} - a*[w_1]_{\mathcal{C}} \tag{18}$$

$$[m_3]_{\mathcal{C}} = −0.982 * \begin{bmatrix} 1 \\ 221.545 \\ 563.590 \end{bmatrix} - -1.003 * \begin{bmatrix} 1 \\ 294.397 \\ 511.738 \end{bmatrix} \tag{19}$$

```python
0.734*Matrix([[1],[448.207],[546.595]]) - -1.003*Matrix([[1],[294.397],[511.738]])
```

$$\therefore [m_3]_{\mathcal{C}} = \begin{bmatrix} 1.737 \\ 624.264129 \\ 914.473944 \end{bmatrix} \tag{20}$$

Augmenting $[m_1]_{\mathcal{C}}, [m_2]_{\mathcal{C}}, [m_3]_{\mathcal{C}}$ gives

$$[m]_{\mathcal{C}} = \begin{bmatrix} −1.003 & 1.981 & 1.737 \\ −295.280191 & 592.582411 & 624.264129 \\ −513.273214 & 1092.355816 & 914.473944 \end{bmatrix} \tag{21}$$ 

$$\Big\Downarrow$$

$$[P]_{\mathcal{C} \leftarrow \mathcal{M}} = [ [m_1]_{\mathcal{C}} \hspace{2mm} [m_2]_{\mathcal{C}} \hspace{2mm} [m_3]_{\mathcal{C}}] \tag{22}$$

$$\Big\Downarrow$$

$$[P]_{\mathcal{C} \leftarrow \mathcal{M}} = \begin{bmatrix} −1.003 & 1.981 & 1.737 \\ −295.280191 & 592.582411 & 624.264129 \\ −513.273214 & 1092.355816 & 914.473944 \end{bmatrix} \tag{23}$$
