```python
%pylab inline
from scipy.integrate import odeint
```

```python
G=6.67e-11
me=5.972e24
mmoon=7.348e24 
ke=G*me
km=G*mmoon
r=6.37e6 
phi=0
sigma=0
xe=0
ye=0
dem=3.844e7/1 
xm=dem*cos(sigma)
ym=dem*sin(sigma)
v=12000 
theta= (1*pi)/22.0125 
def orbit(z,t):
    de=sqrt( (  xe - z[0]  )**2 + ( ye - z[1]  )**2 )
    dm=sqrt( (  xm - z[0]  )**2 + ( ym - z[1]  )**2 )
    f1 = (ke*(xe-z[0]))/(de**3) + (km*(xm-z[0]))/(dm**3)
    f2 = (ke*(ye-z[1]))/(de**3) + (km*(ym-z[1]))/(dm**3)
    dxdt=z[2]
    dydt=z[3]
    dudt=f1
    dvdt=f2
    return [dxdt,dydt,dudt,dvdt]
t=arange(0,144857,0.1) #Launch speed increased ~10x
z0=[r*cos(phi),r*sin(phi),v*cos(theta),v*sin(theta)]
X = odeint(orbit,z0,t)
x=X[:,0]
y=X[:,1]

plt.style.use('fivethirtyeight')
rm=1.737e6
figure(figsize=(20,20)) #(8,8)
subplot(111,aspect=1)
gamma = linspace(0,2*pi,500)
fill( r*cos(gamma), r*sin(gamma) , color='cornflowerblue', alpha=.9)
fill(dem + rm*cos(gamma), rm*sin(gamma) ,color='lightgray', alpha=1)
plot(x,y, color='#ff6000')
plot(xe,ye,'ko')
plot(xm,ym,'ko')
xlabel('x,(m)')
ylabel('y,(m)')
title("Oval_Orbit")
plt.savefig('Oval_Orbit_1', dpi = 90)
print("launch Angle is",
    "{:.10f}".format((theta*180)/pi),
     "degrees")
plt.show()
```
